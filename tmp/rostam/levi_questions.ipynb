#### Question 1
What is the aggregate open rate for the sample provided in the “Sends” table? (Open rate defined as total opens over total sends). Aggregate Unsub per Open (UPO) rate from the “Opens” table?

```python
client = bigquery.Client()
```

```python
sql = """SELECT sum(opened)/count(*) as aggregate_open_rate
    FROM `email-propensity-sandbox.emails.sends`
    """
```

```python
query_job = client.query(sql) # API request
result = query_job.to_dataframe()
```

```python
print("aggregate open rate: {:.2%}".format(result['aggregate_open_rate'][0]))
```

```python
sql = """SELECT sum(optout)/count(*) as agg_upo_rate
    FROM `email-propensity-sandbox.emails.opens`
    """
query_job = client.query(sql) # API request
result = query_job.to_dataframe()
print("aggregate unsub per open (UPO) rate: {:.2%}".format(result['agg_upo_rate'][0]))
```

launch_id is equivalent to `opens.campaign_send_dt` based on the data dictionary

```python
sql = """WITH open AS (
    SELECT campaign_send_dt,
        sum(optout)/count(*) as upo_rate
    FROM `email-propensity-sandbox.emails.opens` opens
    GROUP BY 1),
    upo AS (
        SELECT campaign_send_dt,
            sum(opened)/count(*) as open_rate
        FROM `email-propensity-sandbox.emails.sends` opens
        GROUP BY 1)
    SELECT open.campaign_send_dt,
        open_rate,
        upo_rate
    FROM open LEFT JOIN upo 
        ON open.campaign_send_dt = upo.campaign_send_dt"""
query_job = client.query(sql) # API request
result = query_job.to_dataframe()
```

```python
for index, row in result.iterrows():
    print("launch id: {}\t open rate: {:.2%}\t UPO rate: {:.2%}".format(
        row["campaign_send_dt"], row["open_rate"], row["upo_rate"]))
```
